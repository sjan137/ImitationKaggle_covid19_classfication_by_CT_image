```python
list(df1.COVID)
```

```python
sum_COVID = []
ls = [df1.COVID, df2.COVID , df3.COVID , df4.COVID , df5.COVID, df6.COVID , df7.COVID, df8.COVID, df9.COVID , df10.COVID]
for i in ls:
    sum_COVID.append(list(i))
sum_COVID
```

```python
from collections import Counter
import numpy as np
np_pred = np.array(sum_COVID).T

pred = []
for i in range(100):
    cnt = Counter(np_pred[i])
    pred.append(cnt.most_common()[0][0])
pred
```

```python
file_lst = df1.file_name
pred_lst = pred
```

```python
sum(pred_lst)
```

```python
df = pd.DataFrame({'file_name':file_lst, 'COVID':pred_lst})
# df.sort_values(by=['file_name'], inplace=True)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/YEARDREAM/220209_2차 팀 모의 캐글/predictions/prediction_19.csv', index=False)
```
